```python
import matplotlib.pyplot as plt # Principal biblioteca para geração de graficos no Python
import seaborn as sns # Baseado no Matplotlib, e uma biblioteca que facilitara a criação de graficos

import numpy as np 
import pandas as pd
```

 ### Gráficos de Linha

```python
# Dados retirados do IBGE
vAnos = [2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017]
vDivorcios = [171747, 239070, 347583, 341600, 324921, 341181, 328960, 344526, 373216]

#Craindo o gráfico de linhas
sns.lineplot(x= vAnos, y=vDivorcios)

# Editando o grafico de linhas
plt.title('Divórcios consensuais concedidos em 1ª instância ou por escritura', fontsize=14)
plt.xlabel('Anos', fontsize=12)
plt.ylabel('Quantidade de divórcios', fontsize=12)

# Confirgurando a expessurado do eixo
plt.tick_params(axis='both', labelsize=10)


```

### Gráfico de barras

```python
#Criando uma classe dado
class Dado:
    # Metodo construtor para podermos definir o numero de faces do dado
    def __init__(self, numFaceDado): 
        self.numFaceDado = numFaceDado

    # Função para simularmos o lançamento do dado
    def rolar_Dado(self, numLancamentos): 
        return np.random.randint(1, self.numFaceDado + 1, numLancamentos)

# Inserindo o numero de faces no dados
vDado6 = Dado(6) 

# Definindo o numero de vezes que iremos jogar o dado
vResult = list(vDado6.rolar_Dado(10000)) 

# Criando uma variavel para armazenar a frequencia que cade face e escolhida a cada jogada
vFrequencia = [] 

# Criando um contador para verificar a quantidade de vezes que a face e escolhida
for valor in range(1,7):
    freq = vResult.count(valor)
    vFrequencia.append(freq)
    
    
# Confirurando o gráfico de barras
sns.barplot(x=list(range(1,7)),
            y=vFrequencia,)

# Editado o gráfico de barras
plt.title('Dado de 6 Faces\nRolado 10000 Vezes')        
plt.ylabel('Frequência')                     
plt.xlabel('Resultados')                                    

   
```

```python

class Moeda:
    
    def __init__(self,vLado):
        self.vLado = vLado

    def jogarMoeda(self, vJogarMoeda):
          return np.random.randint(1, self.vLado + 1, vJogarMoeda)


vDadosMoeda = Moeda(2)
vResultado = list(vDadosMoeda.jogarMoeda(10000))

vFreq = []

for valor in range(1,3):  
    freq = vResultado.count(valor)
    vFreq.append(freq)
      
      
sns.barplot(x=list(range(1,3)),
            y=vFreq,)

# Editado o gráfico de barras
plt.title('Dado de 6 Faces\nRolado 10000 Vezes')        
plt.ylabel('Frequência')                     
plt.xlabel('Resultados') 

ticks, labels = plt.xticks()
plt.xticks(ticks, labels=['Cara', 'Coroa'])           
    
```

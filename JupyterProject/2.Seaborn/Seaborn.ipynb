# **Seaborn**

O Seaborn é uma biblioteca de visualização de dados do Python baseado no Matlotlib. Ele provê uma interface de alto nivel para construção
de gráficos estaticos atrativos e informativos.

```python
import pandas as pd
import seaborn as sns

# Definir tema
sns.set_theme(style='darkgrid')
```

```python
#Dados
BaseDados = sns.load_dataset('tips')

#Renomeando as colunas

BaseDados.rename( columns={
    'total_bill' : 'Total_Conta',
    'tip' : 'Gorjeta',
    'sex' : 'Sexo',
    'smoker' : 'Fumante',
    'day' : 'Dia_Semana',
    'time' : 'Periodo',
    'size' : 'Pessoas_na_Mesa'
}, inplace=True)

#Visualizador
BaseDados.head()
```

**Gráfico Relplot**

```python
# Gráfico Relplot
sns.relplot(x='Total_Conta', y='Gorjeta', data=BaseDados)
```

```python
# Gráfico Relplot
# Passando outro parametro como classe no gráfico
sns.relplot(x='Total_Conta', y='Gorjeta', data=BaseDados, hue='Sexo')
```

```python
# Gráfico Relplot
# Utilizando Linhas
sns.relplot(x='Total_Conta', y='Gorjeta', data=BaseDados, hue='Sexo', kind='line')
```

**Histplot**

```python
# Gráfico histplot
# Utilizando Linhas
sns.histplot(x='Total_Conta', data=BaseDados)
```

```python
# Gráfico histplot
# Utilizando Linhas, e separnado por fumantes
sns.histplot(x='Total_Conta', data=BaseDados, hue='Fumante')
```

**Barplot**

```python
# Gráfico barplot
# Utilizando Linhas, e separnado por fumantes
sns.barplot(data=BaseDados, x='Sexo', y='Total_Conta', hue='Fumante')
```

```python
sns.pairplot(BaseDados, hue='Sexo')
```
